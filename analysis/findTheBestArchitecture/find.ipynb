```python
# -*- coding: utf-8 -*-
#允许中文注释


import re #正则模块
import os #文件处理模块
import string  #字符串
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open("nohup-cifar10-micro-search.out", "r") #注意此处的转义字符
a=[]
maxAccuracy=0.0
while True:
    line = file.readline()
    if not line:
        break
    
    if line.startswith("val_acc="):
        #print float(line[15:])
        temp=float(line[len("val_acc="):])
        a.append(temp)
        if float(temp)>maxAccuracy:
            maxAccuracy=float(temp)
```

```python
print maxAccuracy
x=np.arange(0.,len(a)/5,0.2)
plt.rcParams['figure.figsize']=(12,8)   #reParams设置图片尺寸
plt.plot(x,a,color='blue',linewidth=1.5,linestyle='-',marker='.',label=r'$accuracy$')
```
