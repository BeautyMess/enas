```python
# -*- coding: utf-8 -*-
#允许中文注释


import re #正则模块
import os #文件处理模块
import string  #字符串
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline




    
```

```python
file = open("nohup.out", "r") #注意此处的转义字符
a=[]
maxAccuracy=0.0
while True:
    line = file.readline()
    if not line:
        break
    
    if line.startswith("test_accuracy: "):
        #print float(line[15:])
        a.append(float(line[15:]))
        if float(line[15:])>maxAccuracy:
            maxAccuracy=float(line[15:])
```

```python
print maxAccuracy
x=np.arange(0.,310,1)
plt.rcParams['figure.figsize']=(12,8)   #reParams设置图片尺寸
plt.ylim(0.80,1)

plt.grid(True)
plt.plot(x,a,color='blue',linewidth=1,linestyle='--',marker='.',label='accuracy')
```
